# Array

* Array is a sequential data structure. 
* An array of size `n` can hold `n` element. 
* First element of an array can be accessed using `0` as array index  starts from 0 and last element can be accessed using `n-1`

**Example**

```python
stock_price = [298,305,320,301,292]
# We can access elements from the index of the array element.
print(stock_price[0])
```

#### I.  What was the price on 3rd day ?
Here's how the memory address will be generated :

stock_price[0] = 0x00500
stock_price[2] = `0x00500` + `(2 * size_of_integer)` <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;= 0x00500 + (2 * 4) <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;= 0x00508

#### II. What day the price was 301?

```python
for i in range(len(stock_price)):
    if stock_price[i] == 301:
        print("On day {} the stock price was 301".format(i+1))
```

#### III. Complexity of printing an array inserting an element at any index,i, using insert() method and, removing an element at index,i, take `O(n)` time complexity. This is because in the worst case, we may have to shift all the elements of the array to accommodate the new element or to fill the gap left by the removed element.

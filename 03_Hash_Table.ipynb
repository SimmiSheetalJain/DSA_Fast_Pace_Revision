Now, we traverse through the array arr and increase the count at hash[i] for arr[i]. 

i.e.,  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; hash[arr[i]]+=1  

Similarly we can do count of letters if we encounter a problem having string as input. There we can use letter positions, 1 to 26. 

We can make an array of size 26 which will have index 0 to 25. An index i-1 will show the position od a letter and we can store respective counts on that index.


```python
# Example
import numpy as np
arr = [1, 2, 1,3,2]
hash_map = np.zeros(max(arr)+1)

for i in range(len(arr)):
    hash_map[arr[i]] += 1

# Fetch count of 2
print("Count of 2 is: ", int(hash_map[2]))
```

## Implementation of Hash Table

### Hash function

```python
def get_hash(key):
    h = 0
    for char in key:
        h+=ord(char)
    return h%100  # Say our list has 100 indecies so we do mode with 100
# Testing the hash function
get_hash('march 6')
```

### Simple implementation of hash table with a method for adding key-value pair.

```python
class HashTable:
    def __init__(self):
        self.MAX = 100
        self.arr = [None for i in range(self.MAX)]
    def get_hash(self, key):
        h = 0
        for char in key:
            h+=ord(char)
        return h%self.MAX
    ## Adding a key, value pair to the hash table
    def add(self, key, value):
        h = self.get_hash(key)
        self.arr[h] = value
```

```python
hashtable = HashTable()
hashtable.add('march 6', 310)
```

```python
hashtable.arr[:11]
```

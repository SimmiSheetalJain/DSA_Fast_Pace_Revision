Now, we traverse through the array arr and increase the count at hash[i] for arr[i].

Similarly we can do count of letters if we encounter a problem having string as input. There we can use letter positions, 1 to 26.

```python
# Example
import numpy as np
arr = [1, 2, 1,3,2]
hash_map = np.zeros(max(arr)+1)

for i in range(len(arr)):
    hash_map[arr[i]] += 1

# Fetch count of 2
print("Count of 2 is: ", int(hash_map[2]))
```

## Implementation of Hash Table

### Hash function

```python
def get_hash(key):
    h = 0
    for char in key:
        h+=ord(char)
    return h%100  # Say our list has 100 indecies so we do mode with 100
# Testing the hash function
get_hash('march 6')
```

### Simple implementation of hash table with a method for adding key-value pair.

```python
class HashTable:
    def __init__(self):
        self.MAX = 100
        self.arr = [None for i in range(self.MAX)]
    def get_hash(self, key):
        h = 0
        for char in key:
            h+=ord(char)
        return h%self.MAX
    ## Adding a key, value pair to the hash table
    def add(self, key, value):
        h = self.get_hash(key)
        self.arr[h] = value
```

```python
hashtable = HashTable()
hashtable.add('march 6', 310)
```

```python
hashtable.arr[:11]
```

### Let's make the functionality of adding and fetching element from hash table the way we do in dictionary

```python
class HashTable:
    def __init__(self):
        self.MAX = 100
        self.arr = [None for i in range(self.MAX)]

    def get_hash(self, key):
        h = 0
        for char in key:
            h += ord(char)
        return h%self.MAX
    ## Adding a key, value pair to the hash table
    def __setitem__(self, key, value):  ## This is a special method in Python that allows us to use the [] operator to set values in our hash table
        hash_index = self.get_hash(key)
        self.arr[hash_index] = value
    
    ## Fetching a value with key
    def __getitem__(self, key):    ## This is a special method in Python that allows us to use the [] operator to get values from our hash table
        h = self.get_hash(key)
        return self.arr[h]
    
    ## Deleting a key, value pair
    def __delitem__(self, key): ## This is a special method in Python that allows us to use the del operator to delete key, value pairs from our hash table
        h = self.get_hash(key)
        self.arr[h] = None
```

```python
# Testing the HashTable class
myhashtable = HashTable()
myhashtable["march 6"] = 310
print(myhashtable['march 6'])
```

```python
# Adding more key, value pairs to the hash table
myhashtable['march 1'] = 120
myhashtable['dec 13'] = 110
myhashtable['may 6'] = 114
myhashtable['apr 10'] = 105
```

```python
# Out HashTable class is subscriptable, so we can use the [] operator to get values from our hash table
myhashtable['apr 10']
```

```python
# Deleting a key, value pair from our hash table
del myhashtable['apr 10']
```

```python
# No longer key, value pair for 'apr 10' in our hash table exists
if myhashtable['apr 10'] == None:
    print("Key not found")
else:
    print(myhashtable['apr 10'])
```

## Collision Handling in Hash Table

When an index is already occupied by another key and another key is trying to occupy the same index, there occurs a collision. 

In above figure, hash value of 'march 17' will override the value of 'march 6' if we don't handle the collision. 

To handle collision, we have different mechanism. We will try to understanding one of these mechanism, chaining.

In chaining, instead of storing an individual value, we will store a list or linked list.

The second such approach to handle collision is **"Linear Probing"**.  

We linearly probe or search for an empty slot if an index for a key is already occupied.

## Chaining Collision Mechanism Implementation 

```python
## Updated HashTable class to handle collision 
class HashTable:
    def __init__(self):
        self.MAX = 10 # Imagine we have just 0 to 9 indecies in out hash table
        self.arr = [[] for i in range(self.MAX)]
    def get_hash(self, key):
        h = 0
        for char in key:
            h+=ord(char)
        return h%self.MAX
    
    ## Adding a key-value pair

    def  __setitem__(self, key, value):
        h = self.get_hash(key)
        found = False
        for idx,element in enumerate(self.arr[h]):
            if element[0] == key and len(element) == 2:
                self.arr[h][idx] = (key, value)  # Update the value if the key already exists
                found = True
                break
        if not found:
            self.arr[h].append((key, value))  # Add the key-value pair if the key doesn't exist
    
    ## Fetching a value with key
    def __getitem__(self, key):   
        h = self.get_hash(key)
        return self.arr[h]
    
    ## Deleting a key, value pair
    def __delitem__(self, key):
        h = self.get_hash(key)
        self.arr[h] = None
```

```python
stock_prices = HashTable()
stock_prices['march 6'] = 310   
stock_prices['march 7'] = 340
stock_prices['march 12'] = 380
stock_prices['march 17'] = 459
stock_prices.arr
```

## Stack

* Stack works on LIFO (Last in first out) principle
* It is a linear data structure
* Usecases:  
&nbsp;&nbsp;&nbsp;&nbsp;1. When we visit a website, clicking on few links on the websites take us to next page and when we press back button we 
are taken back to previous page. This function in the browser is implemented using stack data structure.  
&nbsp;&nbsp;&nbsp;&nbsp;2. Undo operation in an application is also implemented using stack.  
&nbsp;&nbsp;&nbsp;&nbsp;3. Function calling an a programming language is also performed using stack.  
* Push/pop operation are performed to add/remove an element on/from the top of the stack. This takes constant time, i.e. O(1).
* Seaching an element by value in a stack taken O(n).  

In python a stack can be implemented through,  
&nbsp;&nbsp;&nbsp;&nbsp; 1. list  
&nbsp;&nbsp;&nbsp;&nbsp; 2. Collections.deque  
&nbsp;&nbsp;&nbsp;&nbsp; 3. queue.LifoQueue  

```python
# Stack in python using queue data structure
from collections import deque
stk = deque()
stk.append(34)
stk.append(45)
print(stk)
print(stk[-1]) #The recently added element is top element
```

##### **Why are we using `deque()`? Why can we not use a list to implement a stack?**

A list can be used to implement a stack, but it is not always efficient in terms of memory management.

Suppose the initial capacity of a stack implemented using a list is 10. When an 11th element is inserted, the list needs to allocate additional memory. Typically, the list increases its capacity by resizing (often doubling its size), so it may grow from 10 to 20 memory blocks.

This resizing requires allocating new memory and copying existing elements, which introduces additional overhead.

As shown in the figure, when the stack reaches its initial capacity (10 elements), inserting a new element (139) requires resizing.

A new larger memory block is allocated (capacity increases to 20), and all existing elements are copied from the old memory location to the new one. After copying, the new element (139) is added as the 11th element.

This process of resizing and copying introduces additional time and memory overhead.

So, the recommended class to implement a stack is `collections.deque(iterable[,maxlen])`

## Stack implementation using class and deque() 


```python
from collections import deque

class Stack:
    def __init__(self):
        self.container = deque()
    def push(self, val):
        self.container.append(val)
    def pop(self):
        return self.container.pop()
    def peek(self):
        return self.container[-1]
    def is_empty(self):
        return len(self.container) == 0
    def size(self):
        return len(self.container)
    
```

## Testing our Stack class

```python
stk = Stack()

# Push operation
stk.push(10)
stk.push(20)
stk.push(30)

# Peek Operation
print("Top element: ",stk.peek())

# Pop operation
print("Popped element: ",stk.pop())
print("Top element after pop operation: ",stk.peek())

# Is stack empty ?
print("Is stack empty ?",end="=>  ")
print("Yes") if stk.is_empty() else print("No")
```
